```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-1,1,50)
uL=1.
uR =0.5
T_end = 1

xxx = np.linspace(-1,1,1000)
plt.figure(figsize=(3,4))
plt.plot(xxx, uL*(xxx<0) + uR*(xxx>=0), 'k', lw=2)
plt.xlabel("x")
plt.ylabel("u")
plt.title("Initial condition for Burgers'")
plt.tight_layout()
plt.savefig("burgers_initial_condition_shock.png")
plt.show()

plt.figure(figsize=(8,4))
ss = (uL+uR)/2
for x in xx:
    s = uL*(x<0) + uR*(x>=0)
    t_shock = np.minimum(x/(ss-s),T_end)
    x_end = x + s*t_shock
    plt.plot([x,x_end],[0,t_shock],'k')

# plt.plot([0,0.5],[0,T_end],'b',lw=2)
plt.plot([0,0.75],[0,T_end],'r',lw=3)
# plt.plot([0,1],[0,T_end],'b',lw=2)

plt.xlim(-1,1)
plt.xlabel("x")
plt.ylabel("t")
plt.title("Characteristics curves of a shock wave for Burgers'")
plt.tight_layout()
plt.savefig("burgers_shock_wave_exact.png")
plt.show()


plt.figure(figsize=(8,4))
for x in xx:
    s = uL*(x<0) + uR*(x>=0)
    plt.plot([x,x+s*T_end],[0,T_end],'k')

plt.plot([0,0.5],[0,T_end],'b',lw=2)
plt.plot([0,0.75],[0,T_end],'r',lw=3)
plt.plot([0,1],[0,T_end],'b',lw=2)

plt.xlim(-1,1)
plt.xlabel("x")
plt.ylabel("t")
plt.title("Characteristics curves of a shock wave for Burgers'")
plt.tight_layout()
plt.savefig("burgers_shock_wave.png")
plt.show()
```
